```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
%matplotlib inline
```

## Load data:

We can use Pandas to easily read and parse small to medium sized datasets.

```python
data = pd.read_csv('data/gpa_gre.csv')
data.head(5)
```

## Getting training and test sets

In this example we will do something funny - we will try to predict the GRE score. This dataset is typically used to predict whether or not a candidate will be admitted based on gre, gpa, and rank. We will come back to that example later.

We want to get a training set that we will use cross validation to select a model with. We are going to put the test set aside. Let me make this perfectly clear - we will only use the test set **once**. We will only use it after we have selected a model. It will give a point estimate of the "generalization error" of our model. But it is also pretty meaningless... Reporting the test set error on its own is sort of like reporting the mean without giving the standard deviation, let alone confidence intervals... but I digress.

```python
X = data["gpa"].values.astype(float) # x values will be the gpa
Y = data["gre"].values.astype(float) # we want to predict gre
N = len(X)

np.random.seed(0) # we will set a random seed so we get the same result for replication purposes
shuffled_indices = np.random.permutation(N)

# always shuffle your data - that's the only way the statistics of machine learning works out
X = X[shuffled_indices]
Y = Y[shuffled_indices] # make sure that the Y values still align with the correct X values!

# split data to be 90% training and 10% test.
split_size = 0.9 # the split size
m = int(split_size*N) # the training sample size, take floor
x_train = X[:m]
y_train = Y[:m]
x_test = X[m:]
y_test = Y[m:]
```

## Visualization:

When doing data science (i.e., using machine learning to solve a domain problem), you should always visualize your data first!

```python
plt.plot( x_train, y_train, 'k.' )
plt.xlabel('gpa')
plt.ylabel('gre')
plt.xlim([2.0,4.1])
plt.ylim([200, 900])
plt.title('gre vs gpa')
```

### Houston, we have a problem...

And that is why visualization is so useful - as we can clearly see, there isn't a strong trend between gpa and gre.... 

### Idea: maybe we can use those other features in some way...
But maybe there is a relationship if we condition on the people that were accepted?

```python
accepted = data['admit'].values
accepted = accepted[shuffled_indices]
a_train = accepted[:m]
a_test = accepted[m:]

# accepted
ax = x_train[a_train==1] 
ay = y_train[a_train==1]

# not accepted
nax = x_train[a_train==0]
nay = y_train[a_train==0]


plt.rcParams['figure.figsize'] = (10, 4) # set figure size

plt.subplot(1,2,1)
plt.plot(ax, ay, 'b.')
plt.xlim([2.0,4.1])
plt.ylim([200, 900])
plt.title('accepted')

plt.subplot(1,2,2)
plt.plot(nax, nay, 'r.')
plt.xlim([2.0,4.1])
plt.ylim([200, 900])
plt.title('rejected')
```

Well, that sort of worked. At least we are finding, visually, what looks like a stronger correlation for the rejected candidates... Let's see what happens when we examine the rank.

```python
rank = data['rank'].values
rank = rank[shuffled_indices]
r_train = rank[:m]
r_test = rank[m:]
unique_ranks = np.sort(np.unique(rank))


plt.rcParams['figure.figsize'] = (20, 4) # set figure size
for i in range(len(unique_ranks)):
    plt.subplot(1, len(unique_ranks), i+1)
    plt.plot( x_train[r_train == unique_ranks[i]], y_train[r_train == unique_ranks[i]], 'k.')
    plt.xlim([2.0,4.1])
    plt.ylim([200, 900])
    plt.title('gre and gpa for students with rank ' + str(unique_ranks[i]))
```

Alright, it looks like we teasing apart trends...

Let's try to combine both accepted and rank, then.

```python
plt.rcParams['figure.figsize'] = (20, 8) # set figure size
for i in range(len(unique_ranks)):
    plt.subplot(2, len(unique_ranks), i+1)
    plt.plot( x_train[np.logical_and(r_train == unique_ranks[i], a_train==1)], 
             y_train[np.logical_and(r_train == unique_ranks[i], a_train==1)], 'k.')
    plt.xlim([2.0,4.1])
    plt.ylim([200, 900])
    plt.title('accepted with rank ' + str(unique_ranks[i]))
    
    plt.subplot(2, len(unique_ranks), len(unique_ranks)+i+1)
    plt.plot( x_train[np.logical_and(r_train == unique_ranks[i], a_train==0)], 
             y_train[np.logical_and(r_train == unique_ranks[i], a_train==0)], 'k.')
    plt.xlim([2.0,4.1])
    plt.ylim([200, 900])
    plt.title('rejected with rank ' + str(unique_ranks[i]))
```

### Interpreting the visualization:

Well, it certainly looks like something is there at least sometimes... But how could we capture this relationship? Can we create a model that will allow us to make better predictions? How could we evaluate that model?

**Notice that I have not touched my test set AT ALL thus far! Never, ever do visualizations with your test set - you will bias your model!**
