# Introduction:

This is the introductory tutorial to machine learning and data science. In this tutorial, we will briefly cover the dichotomy between machine learning and data science. We will then go through an example of how a data scientist might solve a basic problem by using machine learning. 

## Machine Learning vs Data Science

There is a lot of disagreement between definitions of machine learning and data science. For the purposes of this tutorial, I will use the following analogy:

Suppose a carpenter wants to make a chair. He might try stacking piles of wood together, but he will eventually consult the local blacksmith. The blacksmith will take somet ime to puzzle over the problem, then might come back to the carpenter with a hammer, nails, saw, etc. The carpenter can then use those tools to build a chair. However, those tools that the blacksmith could be used to build a table, a house, a deck, etc.

Thus, the carpenter is to the blacksmith as a data scientist is to a machine learning engineer. **Data scientists use the tools built by machine learning engineers to solve particular domain problems.** In particular, data scientists work on the special applications, while machine learning engineers and researchers work on general tools.

There is actually another level of machine learning: research. Machine learning researchers invent new statistical, optimization, and algorithmic tools (i.e., the inventors of the hammer). These algorithms are then put into usable code by machine learning engineers. The lines between machine learning research and engineering are blurry and many of the great researchers are also the great engineers.

## Why focus on machine learning?

Data science is certainly in fashion and has proven to be extremely valuable in industry. However, data scientists are necessarily restricted to solving problems that have already been solved. Progress in data science depends on three things: more data, more computing power, and new tools. While a lot of resources are being poured into data and computing power, there is certainly a shortage of machine learning researchers and engineers.

These tutorials, therefore, will focus on the theory and application of *machine learning.* The goal is to give the tools necessary to give engineers the tools necessary to create their own machine learning libraries and begin thinking about how to turn statistical observations into working, machine learning code.  

Data scientists have a lot to gain from a deep understanding of machine learning. First, data scientists often use machine learning daily. The utility of deeply understanding the tools that they use is self evident. Second, the statistics of machine learning have subtle nuances that have important consequences on how we should interpret results. In particular, cross validation is one of the most commonly used *and* misunderstood tools in data science.

Speaking of: I have asked the following interview question dozens of times and have yet to find a data scientist that can answer this question correctly:

"Suppose you are given a model $f$ (that has already been trained) and a labeled sample from a massive dataset. Can you tell me how well that model will perform on the entire dataset? Create a general method that takes in a sample, a model, anything else you need, and returns the evaluation." (No, cross validation and bootstrapping are not correct answers).  Notice that this isn't a machine learning question, per se, but it should be considered elementary for any serious machine learning researcher or engineer. We will answer and code the solution to this exact question by the end of these tutorials.

## Load data:

We can use Pandas to easily read and parse small to medium sized datasets. We won't get into the details here, but basically it reads CSVs no problem.

```python
data = pd.read_csv('data/gpa_gre.csv')
data.head(5)
```

# Regression:

In this example we will do something funny - we will try to predict the GRE score. This dataset is typically used to predict whether or not a candidate will be admitted based on gre, gpa, and rank. We will come back to that example later.

## Generating training and test sets

We want to get a training set that we will use cross validation to select a model with. We are going to put the test set aside. Let me make this perfectly clear - we will only use the test set **once**. We will only use it after we have selected a model. It will give a point estimate of the "generalization error" of our model. But it is also pretty meaningless... Reporting the test set error on its own is sort of like reporting the mean without giving the standard deviation, let alone confidence intervals.|.. but I digress.

```python
X = data["gpa"].values.astype(float) # x values will be the gpa
Y = data["gre"].values.astype(float) # we want to predict gre
N = len(X)

np.random.seed(0) # we will set a random seed so we get the same result for replication purposes
shuffled_indices = np.random.permutation(N)

# always shuffle your data - that's the only way the statistics of machine learning works out
X = X[shuffled_indices]
Y = Y[shuffled_indices] # make sure that the Y values still align with the correct X values!

# split data to be 90% training and 10% test.
split_size = 0.9 # the split size
m = int(split_size*N) # the training sample size, take floor
x_train = X[:m]
y_train = Y[:m]
x_test = X[m:]
y_test = Y[m:]
```

## Visualization:

When doing data science (i.e., using machine learning to solve a domain problem), you should always visualize your data first!

```python
x_limits = [2.0, 4.1]
y_limits = [200,900]

plt.rcParams['figure.figsize'] = (4, 4) # set figure size
plt.plot( x_train, y_train, 'k.' )
plt.xlabel('gpa')
plt.ylabel('gre')
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.title('gre vs gpa')
```

### Houston, we have a problem...

And that is why visualization is so useful - as we can clearly see, there isn't a strong trend between gpa and gre.... 

### Idea: maybe we can use those other features in some way...
But maybe there is a relationship if we condition on the people that were accepted?

```python
accepted = data['admit'].values
accepted = accepted[shuffled_indices]
a_train = accepted[:m]
a_test = accepted[m:]

# accepted
ax = x_train[a_train==1] 
ay = y_train[a_train==1]

# not accepted
nax = x_train[a_train==0]
nay = y_train[a_train==0]


plt.rcParams['figure.figsize'] = (10, 4) # set figure size

plt.subplot(1,2,1)
plt.plot(ax, ay, 'b.')
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.title('accepted')

plt.subplot(1,2,2)
plt.plot(nax, nay, 'r.')
plt.xlim(x_limits)
plt.ylim(y_limits)
plt.title('rejected')
```

Well, that sort of worked. At least we are finding, visually, what looks like a stronger correlation for the rejected candidates... Let's see what happens when we examine the rank.

```python
rank = data['rank'].values
rank = rank[shuffled_indices]
r_train = rank[:m]
r_test = rank[m:]
unique_ranks = np.sort(np.unique(rank))


plt.rcParams['figure.figsize'] = (20, 4) # set figure size
for i in range(len(unique_ranks)):
    plt.subplot(1, len(unique_ranks), i+1)
    plt.plot( x_train[r_train == unique_ranks[i]], y_train[r_train == unique_ranks[i]], 'k.')
    plt.xlim(x_limits)
    plt.ylim(y_limits)
    plt.title('gre and gpa for students with rank ' + str(unique_ranks[i]))
```

Alright, it looks like we teasing apart trends... Let's try to combine both accepted and rank.

```python
plt.rcParams['figure.figsize'] = (20, 8) # set figure size
for i in range(len(unique_ranks)):
    plt.subplot(2, len(unique_ranks), i+1)
    plt.plot( x_train[np.logical_and(r_train == unique_ranks[i], a_train==1)], 
             y_train[np.logical_and(r_train == unique_ranks[i], a_train==1)], 'k.')
    plt.xlim(x_limits)
    plt.ylim(y_limits)
    plt.title('accepted with rank ' + str(unique_ranks[i]))
    
    plt.subplot(2, len(unique_ranks), len(unique_ranks)+i+1)
    plt.plot( x_train[np.logical_and(r_train == unique_ranks[i], a_train==0)], 
             y_train[np.logical_and(r_train == unique_ranks[i], a_train==0)], 'k.')
    plt.xlim(x_limits)
    plt.ylim(y_limits)
    plt.title('rejected with rank ' + str(unique_ranks[i]))
```

### Interpreting the visualization:

Well, it certainly looks like something is there at least sometimes... But how could we capture this relationship? Can we create a model that will allow us to make better predictions? How could we evaluate that model?

**Notice that we have not touched my test set AT ALL thus far! Never, ever do visualizations with your test set - you will bias your model!**

## Creating a model:

We could use a really simple idea in order to model this data here. We could just train 8 different linear regressions - one for each pair of "accept" and "rank." In particular, we will train a "linear model" for each combination of rank and accepted. Our complete model will then just check each student's rank and acceptance decision in order to select the appropriate model to 

We will use scikit learn's built in linear regression algorithm to create our model.

```python
from sklearn import linear_model

class GreModel:
    def __init__( self, unique_ranks=[1,2,3,4], unique_acceptances=[0,1] ):
        # calculate number of models needed
        self.unique_ranks = np.array(unique_ranks)
        self.unique_acceptances = np.array(unique_acceptances)
        self.num_models = len(unique_ranks) * len(unique_acceptances)
        
        # initialize the regression models
        self.models = []
        for i in range( self.num_models ):
            self.models.append( linear_model.LinearRegression() )

    def selectModel( self, a, r ):
        """
        select the correct model based on acceptance decision and rank.
        """
        model_num = self.unique_ranks[self.unique_ranks == r][0] + self.unique_acceptances[self.unique_acceptances == a][0]*len(self.unique_ranks)
        return self.models[model_num-1]
            
    def fit( self, X, A, R, Y ):
        """
        Fit the model with training data
        
        X - the gpa, numpy array of length N
        A - accepted?, numpy array of length N
        R - rank, numpy array of length N
        Y - gre score, numpy array of length N
        """
        X = X.reshape(len(X),-1)
        Y = Y.reshape(len(Y),-1)
        for r in self.unique_ranks:
            for a in self.unique_acceptances:
                indices = np.logical_and(R == r, A==a)
                self.selectModel( a, r ).fit( X[indices], Y[indices] )
                
    def predict( self, X, A, R ):
        """
        Predict according to our model
        
        X - the gpa, numpy array of length N
        A - accepted?, numpy array of length N
        R - rank, numpy array of length N
        """
        X = X.reshape(len(X),-1)
        Y_hat = np.zeros( len(X) )
        for r in self.unique_ranks:
            for a in self.unique_acceptances:
                indices = np.logical_and(R == r, A==a)
                if len(indices) > 0:
                    predictions = self.selectModel( a, r ).predict( X[indices] )
                    Y_hat[indices] = predictions
        return Y_hat
```

### Training our model:

Let's train our model - we will split up our training set further to fit our model and to validate it (again, we are NOT touching the test set yet!).

*Note: Naming conventions tend to differ here. A lot of people call the test set a "hold-out" set and then refer to the "test set" as part of the validation set. I actually prefer that naming scheme and will use it later on. However I want to make it absolutely clear here without getting bogged down in definitions. We will, however, cover this in detail in future tutorials*

```python
gm = GreModel()

shuffled_indices = np.random.permutation(len(x_train)) #shuffle again for good measure
v_split = int( len(x_train) * .8 )

gm.fit( x_train[:v_split], a_train[:v_split], r_train[:v_split], y_train[:v_split] )
y_hat = gm.predict(x_train[v_split:], a_train[v_split:], r_train[v_split:])

# visualize
plt.rcParams['figure.figsize'] = (4, 4) # set figure size
plt.plot(y_train[v_split:], y_hat, 'b.');
plt.plot([300, 900],[300,900],'k');
plt.xlabel('actual GRE scores');
plt.ylabel('predicted GRE scores');
plt.title('predicted versus actual values');

# calculate squared error
mean_squared_error = np.mean( np.power(y_hat - y_train[v_split:], 2) )
mean_absolute_error = np.mean( np.abs(y_hat - y_train[v_split:]) )
print("The mean squared error is: " + str(mean_squared_error))
print("The mean absolute error is: " + str(mean_absolute_error))
print("The standard deviation of GRE scores is: " + str(np.std(y_train[v_split:])))
```

### Interpreting the figure and the model fit

This is a typical figure for analyzing the performance of a regression model. The x axis has all the actual GRE scores, while the predictions are on the y axis. The model would be perfect if all of the points were on the diagonal. Points above the diagonal mean that the model overestimated the GRE score, while a point below the diagonal means that the model underestimated the performance of a student. Visually, we can see that this  model did not do very well.

The mean squared error is very large as well, though we might argue that the mean absolute error is easier to interpret in this example. Specifically, the mean absolute error is around 90, which means that, on average, the model is about 90 points off from the true GRE score. That's pretty bad, considering that the standard deviation of the GRE scores is only about 116.

That being said, as we saw with the initial visualizations, this doesn't necessarily mean that we've chosen a bad model. Instead, this may be more an issue of noisy data - it may be the case that GPA, acceptance, and rank are simply not all that great at predicting GRE scores. This is exactly why we visualized our data in the first place - we can easily see that the data is "noisy" (more on that later). In particular it means that the so-called "Bayes error" is probably quite large. We will discuss all of these concepts in more detail later, but for now just consider the idea that it is impossible to have very good predictions. (Imagine if we didn't visualize our data in the first place - we might have spent a great deal of time guess and testing different models.

Digging deeper: can we use the estimate of the mean squared error to say anything about how well the model will do in general? The short answer is not really. The validation test error, on its own, only gives a point estimate of how badly the model does on our training data. That says nothing about our test (hold out) data, let alone the population in general.
